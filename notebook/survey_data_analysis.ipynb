```python
def number_to_int(number):
    if number != 'None':
        if number == '5 or more':
            return int(5)
        else:
            return int(number)
    else:
        return np.NaN
```

```python
for i in survey_2011.index:
    df = survey_2011.loc[i]
    dataid = df.dataid
    primary = df.primary_residence
    
    year_moved = df.year_moved                                                                         
    workday = df.workday                                                            
    home_working_hours = df.home_working_hours
    num_males = df.males
    num_females = df.females
    num_males = number_to_int(num_males)
    num_females = number_to_int(num_females)
    num_occupants = int(np.nan_to_num(num_females)+np.nan_to_num(num_males))
    
    print(i, dataid, primary, year_moved, workday, home_working_hours, num_males, num_females, num_occupants)
    
    static_feature[year][dataid] = {}

```
